```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

```

```python
df=pd.read_csv("tips.csv")
df
```

```python
df.info()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x="sex",data=df,palette="viridis")
plt.title("Count male and female")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(x="total_bill",data=df,kde=True)
plt.title("Distribution on total bill")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="sex",y="total_bill",data=df)
plt.title("bill distribution by Gender")
plt.show()
```

```python
#piechart
days=df["day"].value_counts()
#piechart
plt.figure(figsize=(10,8))
plt.pie(days,labels=["Sun","Sat","Thus","Fri"],
        autopct="%1.1f%%",colors=["lightcoral","lightgreen","red","blue"],
        startangle=180,explode=(0.05,0.05,0.05,0.05))
plt.title("Day Details")
plt.show()
```

```python
sns.kdeplot(data=df["tip"].dropna(),color="skyblue",shade=True)
plt.title("distribution of tips(kernel density Estimate)")

plt.show()
```
