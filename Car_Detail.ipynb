```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
```

```python
df=pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
df
```

```python
df.info()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x="seller_type",data=df,palette="viridis")
plt.title("selling type count plot")
plt.xlabel("seller type")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.histplot(df["year"].dropna(),bins=30,kde=True,color="yellow")
plt.title("Hisplot for year")
plt.xlabel("year")
plt.ylabel("count")
plt.show()
```

```python
#piechart
cars=df["owner"].value_counts()
#piechart
plt.figure(figsize=(10,8))
plt.pie(cars,labels=["First owner","second owner","third owner","Fourth & above owner","Test Drive Car"],
        autopct="%1.1f%%",colors=["lightcoral","lightgreen","red","blue","royalblue"],
        startangle=180,explode=(0.05,0.05,0.05,0.05,0.05))
plt.title("Car  Owners Details")
plt.show()
```

```python
cars=df["owner"].value_counts
cars
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="seller_type",y="selling_price",data=df,palette="viridis")
plt.title("Boxplot for selling")
plt.xlabel("seller type")
plt.ylabel("selling price")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x="year",y="km_driven",data=df,hue="owner",style="seller_type",palette="rainbow")
plt.title("year vs km_driven eith owner and seller_type")
plt.xlabel("year")
plt.ylabel("km_driven")
plt.legend(title="about car")
plt.show()
```

```python
num=df[["year","selling_price","km_driven"]].corr()
num
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(num,annot=True,cmap="rainbow",center=0)
plt.title("correlation Hearmap of car Dataset")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.violinplot(x="fuel",y="year",hue="seller_type",data=df,split=True,palette="Set2")
plt.title("seller_type distribution by yera and fual")
plt.xlabel("fuel")
plt.ylabel("year")
plt.legend(title="Survived")
plt.show()
```

```python
sns.kdeplot(data=df["year"].dropna(),color="skyblue",shade=True)
plt.title("year Distribution(kernel density Estimate)")
plt.xlabel("year")
plt.ylabel("Density")
plt.show()
```

```python
g=sns.FacetGrid(df,col="seller_type",row="owner",margin_titles=True)
g.map(sns.histplot,"year",bins=20,color="teal")
g.fig.suptitle("age distribution by class and survival",y=1.03)
plt.show()
```

```python
sns.pairplot(df[["year","selling_price","km_driven","transmission"]],hue="transmission",palette="husl",plot_kws={"alpha":0.6},diag_kind="kde")
plt.suptitle("Pair plot of titanic variables",y=1.02)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.swarmplot(x="fuel",y="km_driven",hue="transmission",data=df,palette="Set1",dodge=True)
plt.title("transmission Distribution by fuel and km_driven(Swarm plot)")
plt.xlabel("fuel")
plt.ylabel("km_driven")
plt.legend(title="transmission")
plt.show()
```
